```python
# Check for missing values

for i in df.columns: 
    if df[i].dtype=="object":        
        m=df[i].mode()[0]
        df[i]=df[i].fillna(m)
    else:
        me=df[i].mean()
        df[i]=df[i].fillna(me)
```
